`(2)` 아래와 같은 measurable space를 고려하자.

-   $\Omega=\{a,b,c,d\}$
-   ${\cal F} =\sigma({\cal A})$ where ${\cal A} = \{\{a\}\}$.

아래와 같은 function $X:\Omega \to A:=\{1,2,3,4\}$,
$Y:\Omega \to B:=\{1,2\}$을 고려하자.

-   $X(a)=1, X(b)=2, X(c)=3, X(d)=4$
-   $Y(a)=1, Y(b)=2, Y(c)=2, Y(c)=2$

아래의 물음에 답하라.

-   $X$는 $(\Omega,{\cal F}) \to (A,2^{A})$ 인가? 즉 $X$는
    $(\Omega,{\cal F})$ 에서의 확률변수인가?
-   $Y$는 $(\Omega,{\cal F}) \to (B,2^{B})$ 인가? 즉 $Y$는
    $(\Omega,{\cal F})$ 에서의 확률변수인가?

(풀이)

**$X$는 확률변수가 아님**

집합 $\{2\} \subset 2^A$에 대하여
$\{\omega: X(\omega) \in \{2\}\}=\{b\} \not \in \sigma({\cal A})$ 이므로
$X$는 확률변수가 아님

**$Y$는 확률변수임**

$2^B = \{\emptyset,\{1\},\{2\},B\}$ 의 모든 부분집합 $B^\ast$에 대하여
$\{\omega: X(\omega) \in B^\ast\} \in \sigma({\cal A})$ 이 성립함.

-   $\{\omega: X(\omega) \in \emptyset\} = \emptyset \in \sigma({\cal A})$
-   $\{\omega: X(\omega) \in \{1\}\} = \{a\} \in \sigma({\cal A})$
-   $\{\omega: X(\omega) \in \{2\}\} = \{b,c,d\} \in \sigma({\cal A})$
-   $\{\omega: X(\omega) \in B\} = \{a,b,c,d\} \in \sigma({\cal A})$
