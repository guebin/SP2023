```python
import gym 
import numpy as np
import matplotlib.pyplot as plt 
from matplotlib.animation import FuncAnimation
```

```python
plt.rcParams["animation.html"]="jshtml"
```

```python
env = gym.make('LunarLander-v2',render_mode='rgb_array')
```

```python
env.reset()
_frame = env.render()
```

```python
plt.imshow(_frame)
```

```python
frames = []
env.reset()
for _ in range(300):
    frames.extend(env.render())
    env.step(env.action_space.sample())
env.close()
```

```python
fig, ax = plt.subplots()
ax.imshow(frames)
```

```python
fig, ax = plt.subplots()
FuncAnimation(fig,lambda i: ax.imshow(frames[::100][i]),frames=len(frames[::100]))
```
